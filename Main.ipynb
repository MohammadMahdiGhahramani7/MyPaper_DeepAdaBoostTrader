```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.ensemble import AdaBoostRegressor
import tensorflow as tf
from keras import Sequential, layers
from tensorflow.keras.optimizers import Adam
import plotly.graph_objects as go
from XTilda import Memorize
from DeepAdaBoostRegressor import DeepAdaBoostTrader
!unzip DABT.zip
```

```python
path_list = [f"/content/BTC_USDT-{i}.json" for i in ["30m", "1h", "2h", "4h", "8h"]]
path='/content/DABT'

n_estimators = 100
epochs = 100
lr = 0.02
```

```python
model = DeepAdaBoostTrader(path_list=path_list,
                           n_estimators=n_estimators,
                           epochs=epochs, lr=lr)

model.fit(pretrained_model=True, path=path)
y_hat, y_real = model.predict()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=list(range(len(y_real))), y=y_real, name="Real Price"))
fig.add_trace(go.Scatter(x=list(range(len(y_hat))), y=y_hat[:, 0], name="Predicted Price"))

fig.show()
```

```python
def evaluation(y_real, y_hat):

  dir_real = [1 if y_real[i+1] > y_real[i] else 0 for i in range(len(y_real)-1)]
  dir_hat = [1 if y_hat[i+1] > y_hat[i] else 0 for i in range(len(y_hat)-1)]

  print(f"Accuracy: {accuracy_score(dir_real, dir_hat):.3f}")
  print(f"Recall: {recall_score(dir_real, dir_hat):.3f}")
  print(f"Precision: {precision_score(dir_real, dir_hat):.3f}")
  print(f"F1-Score: {f1_score(dir_real, dir_hat):.3f}")
```

```python
evaluation(y_real=y_real, y_hat=y_hat)
```
